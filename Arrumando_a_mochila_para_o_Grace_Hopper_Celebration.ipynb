```python
import random
import math
```

```python
def criar_cromossomo():
  cromossomo = []
  for i in range(5):
    cromossomo.append(random.randint(0,1))
  return cromossomo
```

```python
def iniciar_populacao(tamanho):
  populacao = []
  for i in range(tamanho):
    individuo = criar_cromossomo()
    populacao.append(individuo)
  return populacao
```

```python
def calcular_fitness(cromossomo):
    tabela_pesos = [1, 2, 3, 4, 5]
    resultado = 0
    for i in range(len(cromossomo)):
        multiplicacao = cromossomo[i] * tabela_pesos[i]
        soma = resultado + multiplicacao
        if soma > 10:
            return resultado
        else:
            resultado = soma
    return resultado
```

```python
def calcular_crossover(pai, mae, taxa_crossover):
    numero_sorteado = random.uniform(0, 1)
    if numero_sorteado <= taxa_crossover:
        ponto_de_corte = random.randint(0, len(mae) - 1)
        filho_1 = pai[:ponto_de_corte] + mae[ponto_de_corte:]
        filho_2 = mae[:ponto_de_corte] + pai[ponto_de_corte:]
        return filho_1, filho_2
    else:
        return pai, mae
```

```python
def selecionar_individuo(populacao, fitness):
    fitness_total = sum(fitness)
    num_sorteado = random.randint(0, fitness_total)
    fitness_acumulado = 0
    for i in range(len(populacao)):
        fitness_acumulado += fitness[i]
        if fitness_acumulado >= num_sorteado:
            return populacao[i], fitness[i]
    posicao_do_individuo_selecionado = random.randint(0, len(populacao) - 1)
    return populacao[posicao_do_individuo_selecionado], fitness[posicao_do_individuo_selecionado]
```

```python
def realizar_mutacao(individuo, taxa_mutacao):
    if random.uniform(0, 1) <= taxa_mutacao:
        i = random.randint(0, len(individuo) - 1)
        if individuo[i] == 1:
            individuo[i] = 0
        else:
            individuo[i] = 1
    return individuo
```

```python
def selecionar_populacao(populacao, fitness, taxa_sobrevivencia):
    populacao_sobrevivente = []
    fitness_sobrevivente = []
    numero_de_sobreviventes = int(taxa_sobrevivencia * len(populacao))
    for i in range(numero_de_sobreviventes):
        sobrevivente, fitness_do_individuo_sobrevivente = selecionar_individuo(populacao, fitness)
        populacao_sobrevivente.append(sobrevivente)
        fitness_sobrevivente.append(fitness_do_individuo_sobrevivente)
    return populacao_sobrevivente, fitness_sobrevivente
```

```python
taxa_mutacao = 0.6
taxa_crossover = 0.9
taxa_sobrevivencia = 0.7
numero_de_geracoes = 1000
tamanho_da_populacao = 10
populacao = iniciar_populacao(tamanho_da_populacao)
```

```python
for geracao in range(numero_de_geracoes):
    print(f'Geração {geracao}')
    fitness = []
    for i in range(len(populacao)):
        fitness.append(calcular_fitness(populacao[i]))
    for i, individuo in enumerate(populacao):
        print(individuo, 'Fitness: ', fitness[i])
    populacao_sobrevivente, fitness_sobrevivente = selecionar_populacao(populacao, fitness, taxa_sobrevivencia)
    nova_populacao = []
    for i in range(tamanho_da_populacao // 2):
        pai, fitness_pai = selecionar_individuo(populacao_sobrevivente, fitness_sobrevivente)
        mae, fitness_mae = selecionar_individuo(populacao_sobrevivente, fitness_sobrevivente)
        filho_1, filho_2 = calcular_crossover(pai, mae, taxa_crossover)
        nova_populacao.append(filho_1)
        nova_populacao.append(filho_2)
    for i in range(len(nova_populacao)):
        individuo = nova_populacao[i]
        individuo_mutado = realizar_mutacao(individuo, taxa_mutacao)
        nova_populacao[i] = individuo_mutado
    populacao = nova_populacao
```
